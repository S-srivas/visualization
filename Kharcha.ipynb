```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data/kharcha.csv', parse_dates=['t'], index_col='t')
```

```python
# Draw Plot
def plot_df(df, x, y, title="", xlabel='Date', ylabel='intervals', dpi=100):
    plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:purple')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.xticks(rotation=90)
    plt.show()

plot_df(df, x=df.index, y=df.amount, title='Kharcha YS Arches Trip')  
```
